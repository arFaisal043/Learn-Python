```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

 -- Categorical plots use for Qualitative data.

```python
# Seaborn have own dataset. search on google -> "seaborn load dataset"
df = sns.load_dataset("tips")
df
```

### Count Plots

```python
sns.countplot(df["sex"]) # y axis default
```

```python
sns.countplot(x = df["sex"])
```

```python
sns.countplot(x=df["sex"], hue=df["smoker"])
```

### Bar Plots

in x axis, use Qualitative data and
in y axis, use Quantitative data.

```python
sns.barplot(x=df["sex"], y=df["tip"])
```

```python
sns.barplot(x=df["sex"], y=df["tip"], estimator=np.sum)
```

### Box Plots

in x axis, use Qualitative data and
in y axis, use Quantitative data.

```python
sns.boxplot(x="tip", y="day", data=df)
```

### Violin Plots

```python
sns.violinplot(x="day", y="total_bill", data=df)
```

### Strip Plots

```python
""" 
- A strip plot, also known as an individual value plot or single-axis scatter plot, is a simple graph that displays 
  individual data points along a single axis to show the distribution of a single numerical variable.

"""
```

```python
sns.stripplot(x="day", y="total_bill", data=df)
```

### Swarm Plots

```python
""" 
- A swarm plot is a type of categorical scatter plot used in data visualization to display the distribution of categorical data. 
It is similar to a scatter plot, but instead of overlapping points, the points are adjusted so that they do not overlap, 
giving a clearer view of the distribution of the data.

"""
```

```python
sns.swarmplot(x="day", y="total_bill", data=df)
```

```python
# we can see here, swarm-plot and violin-plot are almost same. we can use both to better understand
```

```python
sns.violinplot(x="day", y="total_bill", data=df)
sns.swarmplot(x="day", y="total_bill", data=df)
```
