```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 5, 11)  # linear-space
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y)
plt.title("X vs Y Graph")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
```

## Sub-Plots in Matplotlib

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)


plt.subplot(2, 2, 2)
plt.plot(x**2, y)


plt.subplot(2, 2, 3)
plt.plot(x, y**2)


plt.subplot(2, 2, 4)
plt.plot(x**3, y)
```

## Control Canvas by Object Oriented Way

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 1, 1,])
axis1.plot(x, y)

axis2 = fig.add_axes([0.3, 0.5, 0.1, 0.3])
axis2.plot(x, y)
```

## Types of Plots:

```python
# scatter plot
plt.scatter(x, y)
```

```python
# Histogram
plt.hist(x)
```

```python
# Box plot
plt.boxplot(x)
```

## Saving Plots

```python
fig = plt.figure()
a = fig.add_axes([0.1, 0.1, 0.8, 0.8])
a.plot(x, y)

plt.savefig("basic plot.png") # will create a png file
```
