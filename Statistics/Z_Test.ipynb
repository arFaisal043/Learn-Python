```python
import numpy as np
from scipy.stats import norm
```

```python
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]  # here are some mistake that is, In Z-Test must have to greater than 30 sample
population_mean = 170
population_std = 3
sample_mean = np.mean(sample)
n = len(sample)
```

```python
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
z_score
```

```python
p_val = 2 * (1 - norm.cdf(abs(z_score))) # cdf = cumulative distribution function
p_val
```

```python
alpha = 0.05

if p_val < alpha:
    print("We have to reject the Null Hypothesis")
else:
    print("We have to Accept the Null Hypothesis")
```
