```python
import numpy as np
from scipy import stats
```

```python
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)
population_mean = 170
```

```python
t_test = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
t_test
```

```python
p_val = 2 * (1 - stats.t.cdf(abs(t_test), df = n-1))
p_val
```

```python
alpha = 0.05
```

```python
if p_val < alpha:
    print("We have to reject the Null Hypothesis")
else:
    print("We have to Accept the Null Hypothesis")
```
