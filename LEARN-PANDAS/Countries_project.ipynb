### Q11: How many country have Republic in their full name?

```python
df.head(8)
```

```python
cnt = 0

def count_republic(str):
    global cnt       # now we can use cnt variable in our function
    if "republic" in str.lower():
        cnt += 1
    return cnt

df["country_long"].apply(count_republic)
print(cnt) 
```
