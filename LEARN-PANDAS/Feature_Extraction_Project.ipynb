```python
# create a function

def extract_episode(str):
    data = ""
    check = False

    for i in str:
        if i == "(":
            check = True
        if check and i not in "()":     #if(check == True and i != "(" and i != ")")
            data += i
        if i == ")":
            check = False
            break
    
    return data
```

```python
df["Episode"] = df["Title"].apply(extract_episode)   # use extract_episode()
df.head()
```

```python
# Now we have convert "Episode" column into Numerical data. remove "eps"

df["Episode"] = df["Episode"].str.replace(" eps", "")  # _eps convert into empty str
```

### Q2: Make a new column for time stamp

```python
df.loc[0]["Title"]
```

```python
def time_extract(str):
    check = False
    ans = ""

    for i in range(len(str)):
        if str[i] == ")":
            for j in range(i+1, i+20):
                    ans += str[j]

    return ans
```

```python
df["Range of Time"] = df["Title"].apply(time_extract)
df.head()
```

### Q3: Extract Total month from Range of Time columns

```python
from datetime import datetime

def get_total_months(date_range):
    try:
        start, end = date_range.split(' - ')
        
        start_date = datetime.strptime(start, '%b %Y')
        end_date = datetime.strptime(end, '%b %Y')
        
        months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month) + 1
        
        return months
    except:
        return 0
```

```python
df["Total Month"] = df["Range of Time"].apply(get_total_months)
df.head()
```
