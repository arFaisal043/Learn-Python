```python
import pandas as pd
import numpy as np
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
```

```python
df = pd.DataFrame(data)
df
```

### Finding Missing Data

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

### Removing The Missing Data

```python
df.dropna()  # Drop rows with null values
```

```python
df.dropna(thresh=3) # remove rows those are not have at least 3 non NaN values
```

```python
data_frame = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5],
    'C': [1, 2, 3, np.nan, np.nan],
}
```
