```python
import pandas as pd
```

```python
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Subject": ["Math", "Math", "Eng", "Eng", "Math"],
    "Marks": [90, 80, 75, 60, 95]
}
df = pd.DataFrame(data)
df
```

### Line Chart:

```python
df.plot(x="Name", y="Marks", kind="line", title="Student Marks")
```

### Bar Chart

```python
df.plot(x="Name", y="Marks", kind="bar", title="Student Marks")
```

### Histogram

```python
df["Marks"].plot(kind="hist", title="Marks Distribution")
```

### Pie chart

```python
df.plot(kind='pie', y='Marks') 
```

### Box plot

```python
df.plot(kind='box')  
```
