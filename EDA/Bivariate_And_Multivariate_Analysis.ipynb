```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

## Bivariate Analysis:

#### 1. Scatterplot (Numerical - Numerical)

```python
# Scatter plot best for Relationship/Correlation Analysis

sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
```

## Multivariate: (Multiple variable)

```python
# Numerical - Numerical - Categorical

sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'])
```

```python
# 5 variable
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker'], size=tips['size'])
```

##### 2. Bar Plot (Num - Cat)

```python
titanic.head()
```

```python
# Bivariate
sns.barplot(y=titanic['Age'], x=titanic['Pclass'], palette="bright")
```

```python
# Multivariate
sns.barplot(y=titanic['Age'], x=titanic['Pclass'], palette="bright", hue=titanic['Sex'])
```

##### 3. Box Plot (Num - Cat)

```python
# Bivariate
sns.boxplot(x=titanic['Sex'], y=titanic['Age'])
```

```python
# Multivariate
sns.boxplot(x=titanic['Sex'], y=titanic['Age'], hue=titanic['Survived'])
```

##### 4. Displot (Num - Cat)

```python
sns.displot(x=titanic['Age'], hue=titanic['Survived'], kde=True, stat='probability')
```

```python
# Remove histogram
sns.kdeplot(
    data=titanic,
    x='Age',
    hue='Survived',
    common_norm=False,      # Normalize each group separately
    alpha=0.5,               # Transparency
)
```

##### 5. HeatMap (Cat - Cat)

```python
titanic[titanic['Survived'] == 1]['Pclass'].value_counts() 
```

```python
# create a contingency table
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100
```

##### 6. ClusterMap (Cat - Cat)

```python
pd.crosstab(titanic['SibSp'], titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```
