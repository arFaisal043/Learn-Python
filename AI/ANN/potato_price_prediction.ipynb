```python
import pandas as pd
import numpy as np
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("Potato_Price_Prediction_Data.csv")
df.head()
```

```python
df.shape
```

```python
X = df[['potato_kg']] # dataframe (2D)
Y = df['potato_price'] # series (1D)
```

```python
# Scale the data (THIS FIXES THE NaN ISSUE)
scaler_X = MinMaxScaler()
scaler_Y = MinMaxScaler()
X = scaler_X.fit_transform(X)
Y = scaler_Y.fit_transform(Y.values.reshape(-1, 1))
```

### Split Dataset fro Train and Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2) # 20% for test. so 80% for training
```

### ANN Model Building
- Dense(1) в†’ One neuron in the output layer
- input_shape=[1] в†’ Expects 1 input feature (potato_kg)
- activation='linear' в†’ Output = input Г— weight + bias 
- The model will find the best weight (slope) and bias (intercept) during training

- This is a straight line model: y = mx + b where x = potato_kg and y = predicted price.

```python
X_test
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1], activation='linear')
])
```

### Compile The ANN Model

- optimizer='sgd':
SGD = Stochastic Gradient Descent
Controls how the model learns (updates weights)
Works by taking small steps downhill to minimize the loss

- loss='mean_squared_error':
MSE = Average of (predicted - actual)ВІ
Measures how wrong the model is
Lower = better predictions
Standard for regression problems (predicting continuous numbers like price)

```python
model.compile(
    optimizer='sgd', 
    loss='mean_squared_error', 
    metrics=['mae'] # Mean Absolute Error
)
```

### Train Model

```python
model.fit(X_train, Y_train, epochs=50)
```

### Prediction

```python
model.predict(X_test)
```
